Find the Gender in graph, how many Girls and boys are there....

```python
# Distribution status in graph..

ax = sns.countplot(x = 'Gender',data = df)
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
# Gender Distribution in value....

Gender_count = df['Gender'].value_counts(normalize = True)
```

```python
Gender_count
```

```python
# Groupby method use for check the gender contribute amount in sales.. (in values)

df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by ='Amount', ascending = False)
```

```python
# Groupby method use for check the gender contribute amount in sales..(in graph)

sales_Gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by ='Amount', ascending = False)
sns.barplot(x = 'Gender', y = 'Amount', data = sales_Gen)

```

```python
df.columns
```

```python
# Check the age group--- which age_group 

age = sns.countplot(data = df, x = 'Age Group', hue  = 'Gender')
age.bar_label(age.containers[0])
plt.title("Age_ Group based on Gender ")
```

```python
# Total Amount vs Age Group....

age_group = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Age Group', ascending  = False)
sns.barplot(x = 'Age Group', y = 'Amount', data = age_group)
```

```python
 # States
```

```python
plt.figure(figsize = (16,5))
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.barplot(data = sales_state , x = 'State' , y = 'Orders')
```

```python
# Total amount/ sales from top 10 states

plt.figure(figsize = (16,5))
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.barplot(data = sales_state , x = 'State' , y = 'Amount')
```

```python
plt.figure(figsize = (8,4))
ax = sns.countplot(data =df, x = 'Marital_Status')
ax.bar_label(ax.containers[0])
plt.title("Marital Status")

```

```python
df.columns
```

```python
# Marital status based on Gender..

sales_state = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.barplot(data = sales_state , x = 'Marital_Status' , y = 'Amount', hue = 'Gender')
```

```python
# most of the buyers are working in IT sector and Healthcare sector

plt.figure(figsize = (20,5))
sns.countplot(x = 'Occupation', data = df)
```

```python
# From the below graph we can see that most of the buyers are working in IT and helthcare  sector...

plt.figure(figsize = (20,5))
sales_state = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.barplot(x = 'Occupation', y = 'Amount', data = sales_state)
```

```python
#  Most of the customers are order clothes and Food..

plt.figure(figsize = (22,5))
ax = sns.countplot(x = 'Product_Category', data  = df)
ax.bar_label(ax.containers[0])
plt.title("Product category Order")
plt.show()
```

```python
# Product category order based on amount...

plt.figure(figsize = (22,6))
sales_order = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending  = False).head(10)
sns.barplot(x = 'Product_Category', y = 'Amount', data  =sales_order)
```
