```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("Zamoto_Data.xlsx")
```

```python
sns.countplot(x = df['listed_in(type)'])
plt.xlabel("Type of Resturent")
```

```python
df.head()
```

```python
grouped_data = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c = "green", marker = "o")
plt.xlabel("Type of resturent", c = "red", size = 20)
plt.ylabel("votes", c = "red", size = 20)
```

```python
plt.hist(df['rate'], bins = 5)
plt.title("Rating Distribution")
plt.show()
```

```python
sns.barplot(x = df['votes'])
plt.xlabel("Rating distribution")
```

```python
#Average order spending by couples.

couple_data = df['approx_cost(for twp people)']
sns.countplot(x = couple_data)
```

```python
df.head()
```

```python
#which mode (online/offline) has recieved the maximum rating.

plt.figure(figsize = (6,6))
sns.boxplot(data = df, x = 'Online_order', y = 'rate')
```

```python
df.head()
```

```python
pivot_table = df.pivot_table(index='listed_in(type)', columns='Online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```
